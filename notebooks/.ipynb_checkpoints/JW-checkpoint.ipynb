From this sneak-peak of the raw data, we can see an overall upward trend in value over the time range. Let's try to visualize the entire history covered by the dataset:

Looking at this graph, I see the dates are backwards. Let's investigate the type of the "Date" column.

Now let's look at the second dataset:

```python
df2 = pd.read_csv("../data/crypto-markets.csv")
df2
```

This dataset is monolithic; it contains market history for a vast array of coins. I only want to know about bitcoin, so I will filter the rest out.

```python
df2 = df2[df2['symbol']=="BTC"]
df2
```

This dataset still has many attributes we don't care about. Let's prune off everything that isn't included in the first dataset.

```python
df2 = df2.loc[:, ['date', 'open', 'high', 'low', 'close']]
df2
```

Just like the last dataset, let's investigate the type of the "Date" column, and parse it into a better format if need be. 

```python
df2.dtypes
```

```python
df2['date']  = pd.to_datetime(df2['date'])
df2.plot(x='date')
```

It appears we are witnessing the same spike at the the beginning of 2018! Finally, let us try to combine the data. To do so, I'll need to further prune the columns, as well as rename them, so that the dataframes match.

```python
df = df.loc[:, ['Date', 'Open', 'High', 'Low', 'Close']]
df2.columns =  ['Date', 'Open', 'High', 'Low', 'Close']
df3 = pd.concat([df, df2])
df3
```

```python
df3.plot(x='Date')
```

It seems that the datasets differ on their y-axis scale. The first dataset specifically claims to be price in USD, while the second only says "price". Perhaps it is with respect to a different currency. Regardless, since the goal is just to identify broad trends, and the shape of the graphs seems to otherwise be consistent, let's try to normalize the second with respect to the first. To do so, I'll linearly scale the second dataset such that the two datasets have the same min and max in the interval they overlap.

```python
df2_overlapping = df2[(df2['Date'] >= df['Date'].min()) & (df2['Date'] <= df['Date'].max())]

def normalize(v):
    min1 = df['Open'].min()
    max1 = df['Open'].max()
    len1 = max1-min1
    min2 = df2_overlapping['Open'].min()
    max2 = df2_overlapping['Open'].max()
    len2 = max2-min2
    return (v - min2)*((len1)/(len2))+min1

df2_scaled = df2.copy()
df2_scaled.loc[:,['Open','High','Low','Close']] = df2_scaled.loc[:,['Open','High','Low','Close']].apply(normalize)
df2_scaled
```

```python
df4 = pd.concat([df, df2_scaled])
df4.plot(x='Date')
```

Now the data appears to be in agreement. We now have a single graph with a rich history of Bitcoin price trends, particularily surrounding the 2018 spike. However, the data is only useful in its general shape; the y values are clearly unasable. The values of the second dataset were drastically altered to match the scale of the first. In fact, a large portion of the data shows a negative selling price. Still, the fact that the datasets were able to match under a simple linear scaling shows that the datasets generally agree on the trends in the history of Bitcoin pricing.
